```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

from datasets import load_multitask_data
```

```python
sst_train_data, num_labels, para_train_data, sts_train_data = load_multitask_data(
    "data/ids-sst-train.csv", "data/quora-train.csv", "data/sts-train.csv", split ='train')

sst_dev_data, num_labels,para_dev_data, sts_dev_data = load_multitask_data(
    "data/ids-sst-dev.csv", "data/quora-dev.csv", "data/sts-dev.csv", split ='train')
```

```python
para_train_labels = []
for i in para_train_data:
    para_train_labels.append(i[2])
    
sts_train_labels = []
for i in sts_train_data:
    sts_train_labels.append(i[2])
    
sst_train_labels = []
for i in sst_train_data:
    sst_train_labels.append(i[1])
    
para_dev_labels = []
for i in para_dev_data:
    para_dev_labels.append(i[2])
    
sts_dev_labels = []
for i in sts_dev_data:
    sts_dev_labels.append(i[2])
    
sst_dev_labels = []
for i in sst_dev_data:
    sst_dev_labels.append(i[1])
```

```python
plt.subplots(nrows=1, ncols=2, figsize=(20, 6))
plt.subplot(1, 2, 1) # row 1, col 2 index 1
sns.countplot(x=para_train_labels)
plt.title("Train")
plt.ylabel('Count')
plt.xlabel('Classes')

plt.subplot(1, 2, 2) # index 2
sns.countplot(x=para_dev_labels)
plt.title("Dev")
plt.ylabel('Count')
plt.xlabel('Classes')
plt.show()

# i think we should add counts on top of each plot
```
